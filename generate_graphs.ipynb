```
paths = ["/content/drive/My Drive/simulations/no_policy/susceptible/automaton.csv", 
         "/content/drive/My Drive/simulations/no_policy/exposed/automaton.csv", 
         "/content/drive/My Drive/simulations/no_policy/cases/automaton.csv", 
         "/content/drive/My Drive/simulations/no_policy/cases/symptomatic/automaton.csv", 
         "/content/drive/My Drive/simulations/no_policy/cases/asymptomatic/automaton.csv", 
         "/content/drive/My Drive/simulations/no_policy/self-isolating/automaton.csv", 
         "/content/drive/My Drive/simulations/no_policy/recovered/automaton.csv",
         "/content/drive/My Drive/simulations/no_policy/deaths/automaton.csv"]

titles = ["Susceptible Hosts for No Lockdown Policy",
          "Exposed Hosts for No Lockdown Policy",
          "Cases for No Lockdown Policy",
          "Symptomatic Cases for No Lockdown Policy",
          "Asymptomatic Cases for No Lockdown Policy",
          "Self-Isolating Hosts for No Lockdown Policy",
          "Recovered Hosts for No Lockdown Policy",
          "Deaths for No Lockdown Policy"]

y_labels = ["Susceptible",
           "Exposed",
           "Cases",
           "Symptomatic Cases",
           "Asymptomatic Cases",
           "Self-Isolating",
           "Recovered",
           "Deaths"]

plt.figure(figsize=(23,60))  

subplots = [1, 2, 3, 4, 5, 6, 7, 8]
subplot_labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

for index, path in enumerate(paths):
  df = pd.read_csv(path)
  
  ax = plt.subplot(8,2,subplots[index])

  ax.text(-0.05, 1.1, subplot_labels[index], transform=ax.transAxes,
  fontsize=26, fontweight='bold', va='top', ha='right')
  
  plt.title(titles[index], fontsize=18)

  x = np.array(df.index.values)
  x_new = np.linspace(x.min(), x.max(), 300)

  y = np.array(df['0'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="0")
  plt.ylim(0)

  y = np.array(df['1'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="1")
  plt.ylim(0)

  y = np.array(df['2'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="2")
  plt.ylim(0)

  y = np.array(df['3'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="3")
  plt.ylim(0)

  y = np.array(df['4'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="4")
  plt.ylim(0)

  y = np.array(df['5'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="5")
  plt.ylim(0)

  y = np.array(df['6'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="6")
  plt.ylim(0)

  y = np.array(df['7'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="7")
  plt.ylim(0)

  y = np.array(df['8'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="8")
  plt.ylim(0)

  y = np.array(df['9'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="9")
  plt.ylim(0)

  y = np.array(df['9'])
  spl = make_interp_spline(x, y, k=3)
  y_new = spl(x_new)
  plt.plot(x_new, y_new, label="10")
  plt.ylim(0)

  plt.xticks(list(df.index))
  plt.locator_params(axis='x', nbins=37)
  plt.xlabel("Weeks", fontsize=16)
  plt.ylabel(y_labels[index], fontsize=16)

  legend = plt.legend()

plt.show
```
